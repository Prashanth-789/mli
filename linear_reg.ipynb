```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
col_names=['x','y']
data = pd.read_csv('Food-Truck(For Linear Regression Program).csv',names=col_names)
```

```python
data.head()
```

```python
# X=ft['a']
# Y=ft['b']

X = data['x'].values
Y = data['y'].values
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
mean_x=sum(X)/len(X)
mean_y=sum(Y)/len(Y)
print(mean_x)
print(mean_y)
```

```python
sm=0
den=0
for i in range(len(X)):
    sm+=(X[i]-mean_x)*(Y[i]-mean_y)
    den+=(X[i]-mean_x)**2
m=sm/den
print(m)
```

```python
c=mean_y-m*mean_x
```

```python
print(c)
```

```python
line=[m*i+c for i in X]
plt.plot(X,line,label="line fit",color="r")
plt.scatter(X,Y,label="points")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
SSE = sum((Y - line)**2)
SSR = sum((line - mean_y)**2)
SST = sum((Y - mean_y)**2)
R_squared = 1 - (SSR / SST)
print("Sum of Squares Error (SSE):", SSE)
print("Sum of Squares Regression (SSR):", SSR)
print("Sum of Squares Total (SST):", SST)
print("Coefficient of Determination (R^2):", R_squared)
```
